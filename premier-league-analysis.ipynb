```python
import pandas as pd
df = pd.read_csv(r"C:\Users\jj_jo\Documents\premier_league_analysis_data.csv")
df.head()


df.head()
df["hf"] = pd.to_numeric(df["hf"], errors="coerce")
df["af"] = pd.to_numeric(df["af"], errors="coerce")
```

```python
# rename column names
change_names = {"hometeam":"home team", "awayteam": "away team", "fthg": "FT home goals", "ftag": "FT away goals",
               "ftr": "Match winner", "hthg": " HT home goals", "htag": "HT away goals", "htr": "HT result",
               "hst": "HS on target", "ast": "AS on target", "hf": "home fouls", "af": "away fouls", "hr": "home red card",
               "ar": "away red card", "hs": "home shots", "as": "away shots"}

df = df.rename(mapper = change_names, axis = 1)
df
```

```python
# drop unwanted columns

drop_col = ["hc", "ac", "hy", "ay", "referee"]
df = df.drop(labels = drop_col, axis = 1)
df
```

```python
# what team commits the most fouls
# need to take all home teams and away team name and home fouls and away foul data
#
away_fouls = df.groupby("away team").agg({"away fouls": "sum"}).sort_values(by = "away fouls", ascending = False)
away_fouls.columns = ["total away fouls"]
away_fouls = away_fouls.round(decimals = 0)
away_fouls.reset_index()

# Brighton team with most away fouls
```

```python
# Brighton for team with most home fouls
home_fouls = df.groupby("home team").agg({"home fouls" : "sum"}).sort_values(by = "home fouls", ascending = False)
home_fouls.columns = ["total home fouls"]
home_fouls = home_fouls.round(decimals = 0)
home_fouls.reset_index()
home_fouls
```

```python
# merging home fouls and way fouls into a table

merged_data = pd.merge(left = home_fouls,
        right = away_fouls,
        left_index = True,
        right_index = True,
        how = "outer")

merged_data.sort_values(by = ["total home fouls", "total away fouls"], ascending = [False, False])
```

```python
# Brighton commits the most number of fouls
merged_data["total fouls"] = merged_data["total home fouls"] + merged_data["total away fouls"]
merged_data = merged_data[["total home fouls", "total away fouls", "total fouls"]]
merged_data.sort_values(by = "total fouls", ascending = False)
```

```python
# calculate average red cards of home team
red_cards_home = df.groupby("home team").agg({"home red card": "mean"}) * 100
red_cards_home.columns = ["red card average home"]
red_cards_home = red_cards_home.sort_values(by = "red card average home", ascending = False).reset_index()
red_cards_home.set_index("home team", inplace = True)
red_cards_home.columns





```

```python
# average red cards of away team
red_cards_away = df.groupby("away team").agg({"away red card": "mean"}) 
red_cards_away.columns = ["red card average away"]
red_cards_away = red_cards_away.sort_values(by = "red card average away", ascending = False).reset_index()
red_cards_away.set_index("away team", inplace = True)
red_cards_away.columns
```

```python
# merging red card tables
combined = pd.merge(left = red_cards_home,
        right = red_cards_away,
        left_index = True,
        right_index = True,
        how = "outer")

combined_red_card_data = combined.sort_values(by = ["red card average home", "red card average away"], ascending = False)
combined_red_card_data
```

```python
combined_red_card_data["total red cards average"] = combined_red_card_data["red card average home"] + combined_red_card_data["red card average away"]
combined_red_card_data = combined_red_card_data[["red card average home", "red card average away", "total red cards average"]]
combined_red_card_data

# calculating total red cards
```

```python
df["Match winner"].value_counts()
# home teams win more games than away teams
# group teams by wins



```

```python
# filtering out the team who won so I can store this info in a new column
# next step will be to count the number of times these teams has won
def winning_team(row):
    if row["Match winner"] == "H":
        return row["home team"]
    elif row["Match winner"] == "A":
        return row["away team"]
    else:
        return "Draw"
    
    
df["winning team"] = df.apply(winning_team, axis = 1)
df
```

```python
df 
# grouping by teams and counts the number of times a team has won

total_wins = df.groupby("winning team").agg({"winning team": "count"})
total_wins.columns = ["total wins"]
total_wins = total_wins.sort_values(by = "total wins", ascending = False).reset_index()
             
    

    

    


        
    
```

```python
# merge total_wins and combined_red_card_data
total_wins.columns
total_wins.set_index("winning team", inplace = True)
full_merge = pd.merge(left = combined_red_card_data,
        right = total_wins,
        left_index = True,
        right_index = True,
        how = "outer")

full_merge = full_merge.sort_values(by = ["total red cards average", "total wins"], ascending = [False, False])
full_merge
```

```python
compare_data = full_merge[["total red cards average", "total wins"]]
compare_data.sort_values(by = "total wins", ascending = False)

# number of red cards don't necessarily correlate with winning matches
#teams with a high red card average also have a lot of wins - although teams with higher percentage of red cards average tend 
# to have less wins
# total red cards average is a percentage %
```

```python
# The coach has asked you whether you can predict the outcome of a game by the results at halftime 
# use HT result and compare that to whether they won the game or not
# write function to give us a new column of the winning team at half time

def half_time_winner(row):
    if row["HT result"] == "H":
        return row["home team"]
    elif row["HT result"] == "A":
        return row["away team"]
    
    
    
df["Lead at HT"] = df.apply(half_time_winner, axis = 1)
df
```

```python
half_time_results = df.groupby("winning team").agg({"Lead at HT": "count", "winning team": "count"})
half_time_results.columns = ["num of matches in lead", "number of matches won"]
half_time_results = half_time_results.sort_values(by = "number of matches won", ascending = False)
half_time_results.reset_index()
```

```python
# calculate the number of games each team has played
# number of games at home
total_games_home_team = df.groupby("home team").agg({"home team": "count"})
total_games_home_team.columns = ["total games at home"]
total_games_home_team.reset_index()
```

```python
# calculate total games away for each team
total_games_away_team = df.groupby("away team").agg({"away team" : "count"})
total_games_away_team.columns = ["total games away"]
total_games_away_team.reset_index()
```

```python
# merge the two dataframes of total games at home and away 
merge_total_games = pd.merge(left = total_games_home_team,
        right = total_games_away_team,
        left_index = True,
        right_index = True,
        how = "outer")

merge_total_games = merge_total_games.reset_index()
merge_total_games
```

```python
#add up total games
merge_total_games["total games played"] = merge_total_games["total games at home"] + merge_total_games["total games away"]
merge_total_games = merge_total_games[["home team", "total games played"]]
merge_total_games
```

```python
# need to now merge merge_total_games into half_time_results dataframe - merge on "winning team"  and "home team"
full_data = pd.merge(left = merge_total_games,
        right = half_time_results,
        left_on = "home team",
        right_on = "winning team",
        how = "outer")

rename_col = {"home team": "team", "num of matches in lead": "num of matches in lead at HT"}
full_data = full_data.rename(mapper = rename_col, axis = 1)
full_data = full_data.sort_values(by = "number of matches won", ascending = False).reset_index()
full_data
```

```python
# creating pivot table
pivot_table = pd.pivot_table(full_data, index = "team", values = ["total games played", "num of matches in lead at HT", "number of matches won"])
pivot_table = pivot_table.reset_index()
pivot_table.columns
```

```python
# calculate percentage games won = num of matches in lead at half time / number of matches won multiply by 100
pivot_table["percent of games won with lead"] = pivot_table["num of matches in lead at HT"] / pivot_table["number of matches won"] * 100
pivot_table = pivot_table[["team", "num of matches in lead at HT", "number of matches won", "percent of games won with lead"]]
pivot_table = pivot_table.sort_values(by = "percent of games won with lead", ascending = False)
pivot_table

# shows teams who have the lead at half time are more likely to go on and win the whole game
```

```python
# calculating the average games won if you have the lead at half time 
# results show 64% which shows positive correlation between lead at HT and games won
pivot_table["percent of games won with lead"].mean()
```
